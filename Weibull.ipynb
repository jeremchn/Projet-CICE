```python
import numpy as np
import matplotlib.pyplot as plt

# Weibull distribution parameters
A = 7.1  # Scale parameter (m/s)
k = 2.4  # Shape parameter

# Wind speed range (0 to 20 m/s)
v = np.linspace(0, 20, 500)

# Weibull probability density function (PDF)
def weibull_pdf(v, A, k):
    return (k / A) * (v / A)**(k - 1) * np.exp(-(v / A)**k)

# Compute PDF values
pdf = weibull_pdf(v, A, k)

# Plot Weibull distribution
plt.figure(figsize=(10, 6))
plt.plot(v, pdf, label=f"Weibull PDF (A={A}, k={k})", color="blue")
plt.fill_between(v, pdf, color="blue", alpha=0.2)
plt.title("Weibull Wind Speed Distribution at Hub Height", fontsize=14)
plt.xlabel("Wind Speed (m/s)", fontsize=12)
plt.ylabel("Probability Density", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=12)

# Save the plot
plt.savefig("weibull_wind_distribution.png", dpi=300)

plt.show()
def weibull_cumul (v, A, k):
    return (1-np.exp((-v/A)**k))
# Plot Weibull distribution
plt.figure(figsize=(10, 6))
plt.plot(v, cum, label=f"Weibull Cumul (A={A}, k={k})", color="blue")
plt.fill_between(v, pdf, color="blue", alpha=0.2)
plt.title("Fonction de rpartition", fontsize=14)
plt.xlabel("Wind Speed (m/s)", fontsize=12)
plt.ylabel("Probability", fontsize=12)
plt.grid(True, linestyle="--", alpha=0.6)
plt.legend(fontsize=12)

# Save the plot
plt.savefig("weibull_wind_cumule.png", dpi=300)

plt.show()
```
